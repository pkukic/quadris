```python
import numpy as np
```

```python
START_OFFSET = 81
P_NUMBER = 1
S_NUMBER = 5
B_NUMBER = 2
L_NUMBER = 4
Z_NUMBER = 4
T_NUMBER = 5

P_OFFSET = START_OFFSET
S_OFFSET = P_OFFSET + P_NUMBER
B_OFFSET = S_OFFSET + S_NUMBER
L_OFFSET = B_OFFSET + B_NUMBER
Z_OFFSET = L_OFFSET + L_NUMBER
T_OFFSET = Z_OFFSET + Z_NUMBER
```

```python
def P_NEUTRAL(i, j):
    return [(i, j)]
```

```python
def S_NEUTRAL(i, j):
    return [(i, j), (i, j+1), (i+1, j), (i+1, j+1)]
```

```python
def B_UP_DOWN(i, j):
    return [(i, j), (i+1, j), (i+2, j), (i+3, j)]
```

```python
def B_LEFT_RIGHT(i, j):
    return [(i, j), (i, j+1), (i, j+2), (i, j+3)]
```

```python
def L_UP_LEFT(i, j):
    return [(i, j), (i+1, j), (i+1, j+1), (i+1, j+2)]
```

```python
def L_UP_RIGHT(i, j):
    return [(i, j+2), (i+1, j), (i+1, j+1), (i+1, j+2)]
```

```python
def L_DOWN_LEFT(i, j):
    return [(i+1, j), (i, j), (i, j+1), (i, j+2)]
```

```python
def L_DOWN_RIGHT(i, j):
    return [(i, j), (i, j+1), (i, j+2), (i+1, j+2)]
```

```python
def L_VERT_UP_LEFT(i, j):
    return [(i, j), (i, j+1), (i+1, j+1), (i+2, j+1)]
```

```python
def L_VERT_UP_RIGHT(i, j):
    return [(i, j), (i+1, j), (i+2, j), (i, j+1)]
```

```python
def L_VERT_DOWN_LEFT(i, j):
    return [(i+2, j), (i, j+1), (i+1, j+1), (i+2, j+1)]
```

```python
def L_VERT_DOWN_RIGHT(i, j):
    return [(i, j), (i+1, j), (i+2, j), (i+2, j+1)]
```

```python
def Z_LEFT(i, j):
    return [(i, j), (i, j+1), (i+1, j+1), (i+1, j+2)]
```

```python
def Z_RIGHT(i, j):
    return [(i+1, j), (i+1, j+1), (i, j+1), (i, j+2)]
```

```python
def Z_UP(i, j):
    return [(i, j), (i+1, j), (i+1, j+1), (i+2, j+1)]
```

```python
def Z_DOWN(i, j):
    return [(i, j+1), (i+1, j), (i+1, j+1), (i+2, j)]
```

```python
def T_UP(i, j):
    return [(i, j+1), (i+1, j), (i+1, j+1), (i+1, j+2)]
```

```python
def T_DOWN(i, j):
    return [(i, j), (i, j+1), (i, j+2), (i+1, j+1)]
```

```python
def T_LEFT(i, j):
    return [(i, j+1), (i+1, j), (i+1, j+1), (i+2, j+1)]
```

```python
def T_RIGHT(i, j):
    return [(i, j), (i+1, j), (i+2, j), (i+1, j+1)]
```

```python
def filled_mat(func):
    mat = np.zeros((4, 4))
    for i in range(4):
        for j in range(4):
            if (i, j) in func(0, 0):
                mat[i][j] = 1
    return mat
```

```python
def cell_index(coords):
    i = coords[0]
    j = coords[1]
    return i * 9 + j
```

```python
cell_index((2, 3))
```

```python
cell_index((8, 8))
```

```python
def P_NEUTRAL_ENTRY(i, j, n):
    return list(map(cell_index, P_NEUTRAL(i, j))) + [P_OFFSET + n]
```

```python
functions_and_offsets_dict = {functions[i].__name__: (offsets[i]) for i in range(len(functions))}
```

```python
functions_and_offsets_dict
```

```python
def ROW_ENTRY(func, i, j, n):
    return sorted(list(map(cell_index, func(i, j))) + [functions_and_offsets_dict[func.__name__] + n])
```

```python
ROW_ENTRY(T_UP, 7, 6, 2)
```

```python
P_ROWS = []
for i in range(9):
    for j in range(9):
        for n in range(P_NUMBER):
           P_ROWS.append(ROW_ENTRY(P_NEUTRAL, i, j, n))
```
